## Prerequisite knowledge for assessment

The assessment will require you to find anagrams, palindromes, and palingrams. To do this you need to understand what they are. The below section presents a simple explanation of these concepts.
### 1. Anagrams

An anagram is a play on words created by rearranging the letters of the original word to make a new word or phrase. Anagram examples can be fun and witty, and they often end in hilarious results.
Anagrams are often longer words or phrases that do not necessarily mean anything but are fun to come up with and say. There are also anagrams of simple words that create random, new words that are not relevant to one another.
Single word anagram examples include:
    
* Tar = Rat
* Arc = Car
* Elbow = Below
* State = Taste
* Cider = Cried
* Dusty = Study
* Night = Thing

There are more complex phrases which include multiple words; however, this assessment will focus on anagrams of single words found in the dictionary.

### 2. Palindromes
A palindrome is a word, phrase, number, or sequence of words that reads the same backward as forward. Punctuation and spaces between the words or lettering is allowed. This assessment will focus on single word, and two-word palindromes.
5
Single word palindrome examples include:
* anna
* civic
* kayak
* level
* madam
* mom
* noon
* radar
* redder

### 3. Palingrams
A palingram is a sentence in which the letters, syllables, or words read the same backward as they do forward. Multiple word palindromes (palingrams) include:
* top spot
* my gym
* ala cicala

#### load(file)
Loads a text file as a list.

Arguments:
-text file name

Exceptions:
-IOError if filename not found.

Returns:
-A list of all words in text file in lower case.

Requires-import sys

```python
def load(file):
    """Open a text file & turn contents into a list of lowercase strings."""
    try:
        with open(file) as in_file:
            loaded_txt = in_file.read().strip().split('\n')
            loaded_txt = [x.lower() for x in loaded_txt]
            return loaded_txt
    except IOError as e:
        print("{}\nError opening {}. Terminating program.".format(e, file),
              file=sys.stderr)
        sys.exit(1)
```

#### find_anagrams(word_for_search, search_list_dictionary):
Find dictionary palingrams.

```python
# use for activity A
def find_anagrams(word_for_search, search_list_dictionary):
    """Search list for anagrams of a user-prompted word, and compile list."""
    if word_for_search.isalpha() == True: # user input must consist of alphabetical characters only
        print("Using word = " + str(word_for_search) + "\n")
        anagrams_found = [] # initialize list for compiling anagrams found
        anagrams_found.insert(0, word_for_search) # shows user's word as first item in anagram list
        for dictionary_word in search_list_dictionary:
            if len(word_for_search) == len(dictionary_word): # checks for single word anagrams only
                if sorted(word_for_search) == sorted(dictionary_word):
                    if word_for_search != dictionary_word: # avoids duplicating user's word in list
                        anagrams_found.append(dictionary_word)
                        
        return anagrams_found
    else:      # if user input is invalid
        main() # initial explanation and prompt will be re-displayed
```

#### find_palingrams(dictionary_list)
Find dictionary palingrams.

```python
# use for activity B
def find_palingrams(dictionary_list):
    """Search list for combinations of two-word palingrams, and compile list."""
    dictionary_list.extend(["a", "i"])
    pairs_of_twos = [] # initialize list for compiling palingrams found
    for every_word in dictionary_list:
        for other_word in dictionary_list:
            word_pair_combined = every_word + other_word # ignores space between pairs
            if "".join(reversed(word_pair_combined)) == word_pair_combined:
                if len(word_pair_combined) > 2: # omits "a a" and "i i" as palingrams
                    pairs_of_twos.extend([every_word + " " + other_word])
    pairs_of_twos.sort()
    return pairs_of_twos
    
```

#### print_generic_list(list_items_category_name, list_to_print):
Print a list of the items in any list passed to the function. Use the first parameter to describe the list.

```python
# use for activity A and C
def print_generic_list(list_items_category_name, list_to_print):
    """Print list(s) of anagrams and/or palindromes."""
    if len(list_to_print) == 0: # if the list is empty
        none_message = "There are no {0} to be displayed.\n".format(list_items_category_name) # "anagrams"/"palindromes" (activity A)
        print(none_message)
    else:
        print(list_items_category_name.title()) # "anagrams"/"palindromes" (activity A), or "dictionary palindromes" (activity C)
        for each_item in list_to_print:
            print(each_item)     
```

#### find_palindromes(anagram_list)
Find palindromes in the anagrams list.

```python
# use for activity A and C
def find_palindromes(anagram_list):
    """Find palindromes in a list."""
    palindromes_found = [] # initialize list for compiling palindromes found
    for each_word in anagram_list:
        if "".join(reversed(each_word)) == each_word:
            palindromes_found.append(each_word)
            palindromes_found.sort() # sorts list alphabetically to display for activity C
    return palindromes_found
```

#### print_pairs_of_twos(dictionary_of_pairs):
Print the keywords and values of an dictionary passed to the method.

```python
# use for activity B
def print_pairs_of_twos(dictionary_of_pairs):
    """Print list of two-word palingrams."""
    for pairs in dictionary_of_pairs: 
        print(pairs)
        
```

#### write_anagrams_palindromes_to_file(original_word, anagrams_list, palindromes_list):
Write the lists of anagrams and the palindromes to a file.Use the original word in the file names and descriptions.

```python
# use for activity A
def write_anagrams_palindromes_to_file(original_word, anagrams_list, palindromes_list):
    """Write the lists of anagrams and palindromes to a file."""
    date_time = datetime.datetime.now()
    file_name = f"{original_word}" + date_time.strftime("-%m.%d.%Y_%H.%M.%S") + ".txt" # file name format
    anagrams_message = f"{original_word.upper()}\nTHE ANAGRAMS ARE:\n\n"
    palindromes_message = "\nTHE PALINDROMES IN THE ANAGRAMS COLLECTION ARE:\n\n"
    with open(file_name, "a") as compiled_lists:
        compiled_lists.write(anagrams_message)
        compiled_lists.write("\n".join(anagrams_list)) # writes anagrams from list separated by space 
        compiled_lists.write(palindromes_message)
        if len(palindromes_list) == 0:
            none_message = "There are no palindromes."
            compiled_lists.write(none_message)
        else:
            compiled_lists.write("\n".join(palindromes_list)) # writes palindromes from list separated by space
            
```

#### write_palingrams_to_file(palingram_list):
Write the lists of palingrams to a file.

```python
# use for activity B
def write_palingrams_to_file(palingram_list):
    """Write the lists of palingrams to a file."""
    file_name = "dictionaryPalingrams.txt"
    with open(file_name, "w") as file_to_overwrite:
        file_to_overwrite.write("\n")
        file_to_overwrite.write("\n".join(palingram_list)) # writes palingrams from list separated by space
        
```
